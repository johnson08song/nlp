```python
import jieba.analyse
import pandas as pd

# 读取CSV文件
df = pd.read_csv('/content/drive/MyDrive/手動新增.csv')

# 确保评论栏是字符串类型，并填充NaN为""
df['Comments'] = df['Comments'].astype(str).fillna('')

# 合并所有评论为一个字符串
text = ' '.join(df['Comments'])

# 提取关键词
keywords = jieba.analyse.extract_tags(text, topK=20, withWeight=False)

# 输出关键词和权重
print(keywords)
```

```python
import jieba.analyse

dic = [
    {'comment': '出貨迅速，不常開伙，就買來備用。這價錢很可以，質感也還不錯', 'sentiment': 'neutral'},
    {'comment': '刀柄都有破損 刀子也有刮傷 已申請換貨 請給我全新的商品 出貨請檢查品質', 'sentiment': 'bad'},
    {'comment': '方便購買，方便購買', 'sentiment': 'good'},
    {'comment': '便宜又好用，過年來個大更新', 'sentiment': 'good'}
]


all_comments = ' '.join([i['comment'] for i in dic])
keywords = jieba.analyse.extract_tags(all_comments, topK=20, withWeight=False)
print(keywords[:3])
```

```python
import jieba.analyse
import pandas as pd

# Create DataFrame with the reviews
df = pd.DataFrame([
    ['特價購入 2188 之後竟然還再特價1990 算了不退了'],
    ['運轉聲音真的蠻小的，不過塑膠味也是真的蠻重的'],
    ['好漂亮的氣炸鍋，是買來送朋友的，結果看到實品覺得美的不行！而且好快就到貨～'],
    ['好用，便宜，有質感，出貨也快速'],
    ['超棒的出貨速度，超讚的商品品質。'],
], columns=['review'])

# Extract keywords using jieba
def extract_keywords(text):
    return jieba.analyse.extract_tags(text, topK=5)

# Apply the function to the 'review' column
df['keywords'] = df['review'].apply(extract_keywords)

# Show the DataFrame with extracted keywords
df.head()
```
